```python
import pandas as pd
import numpy as np
```

```python
main_data = pd.read_csv('ready_data.csv')
main_data.columns
```

```python
const = main_data[['constructor', 'GP_name', 'constructor_home', 'const_relaiblity', 'active_const']]
const = const.groupby(['constructor', 'GP_name']).agg({
    'constructor_home': 'first',
    'const_relaiblity': 'max',
    'active_const': 'max'
}).reset_index()
const
```

```python
const.to_csv('constructors.csv', index=False)
```

```python
driver = main_data[['driver', 'GP_name', 'driver_home', 'driver_conf', 'active_driver', 'best_position', 'driver_score', 'dob', 'date_first_race']]
driver = driver.groupby(['driver', 'GP_name']).agg({
    'driver_home': 'first',
    'driver_conf': 'max',
    'active_driver': 'first',
    'best_position': 'max',
    'driver_score': 'max',
    'dob': 'first',
    'date_first_race': 'first'
}).reset_index()
driver
```

```python
driver.to_csv('drivers.csv', index=False)
```
